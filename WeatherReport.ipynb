```python
up = k_df[(k_df['region']=='Kerala')][['location_name','temperature_celsius']]
area_up = up.groupby('location_name')['temperature_celsius'].mean()
up = k_df[(k_df['region']=='Kerala')][['location_name','temperature_fahrenheit']]
area_up_1 = up.groupby('location_name')['temperature_fahrenheit'].mean()
up = k_df[(k_df['region']=='Kerala')][['location_name','humidity']]
area_up_2 = up.groupby('location_name')['humidity'].mean()
# area_up
# area_up_1
# area_up_2
```

```python
km = pd.DataFrame(area_up)
km1 = pd.DataFrame(area_up_1)
km2 = pd.DataFrame(area_up_2)
```

```python

km.rename(columns={'temperature_celsius':'mean_tem_c'},inplace=True)
km['states'] = ['Alappuzha', 'Ernakulam', 'Idukki','Kannur','Kasaragod','Kollam','Kottayam','Kozhikode','Mahe','Malappuram','Palakkad','Pathanamthitta','Thiruvananthapuram','Thrissur']

km1.rename(columns={'temperature_fahrenheit':'mean_tem_f'},inplace=True)
km1['states'] = ['Alappuzha', 'Ernakulam', 'Idukki','Kannur','Kasaragod','Kollam','Kottayam','Kozhikode','Mahe','Malappuram','Palakkad','Pathanamthitta','Thiruvananthapuram','Thrissur']


km2.rename(columns={'humidity':'mean_hum'},inplace=True)
km2['states'] = ['Alappuzha', 'Ernakulam', 'Idukki','Kannur','Kasaragod','Kollam','Kottayam','Kozhikode','Mahe','Malappuram','Palakkad','Pathanamthitta','Thiruvananthapuram','Thrissur']

```

```python
kd = pd.merge(km, km1, on='states', how='inner')
k_data = pd.merge(kd, km2, on='states', how='inner')
k_data
```

```python
# Create a pie chart using Plotly Express
import plotly.express as px
fig = px.pie(k_data, values='mean_tem_c', names='states', title='Mean Temperature(C) For each Districts In Kerala')

# Show the plot
fig.show()
```
